```python
from pandas import DataFrame, read_json

import pandas as pd
import matplotlib.pyplot as plt
import sys

%matplotlib inline

print('Python version ' + sys.version)
print('Pandas version ' + pd.__version__)

business_json_df = pd.read_json('yelp_academic_dataset_business.json', lines=True)
business_json_df.count()

business_json_df.to_csv('C:/Users/revanth/Desktop/business.csv')

filteredBusinesses = business_json_df[business_json_df.attributes.map(lambda x: not pd.isnull(x) and any(y.startswith('Restaurant') or y.startswith('Alcohol') for y in x.keys())) | business_json_df['categories'].astype(str).str.contains("Restaurants|Food|Restaurant|Bars", case=False) |  business_json_df.name.astype(str).str.contains("Restaurants|Restaurant", case=False)]
filteredBusinesses.count()

filteredBusinesses.to_csv('C:/Users/revanth/Desktop/filteredbusiness.csv')

review_json_df = pd.read_json('yelp_academic_dataset_review.json', lines=True)
review_json_df.count()

review_json_df.to_csv('C:/Users/revanth/Desktop/agg_review.csv')

a = pd.read_csv('C:/Users/revanth/Desktop/filteredbusiness.csv')
a.drop(a.columns[a.columns.str.contains('unnamed',case = False)],axis = 1)

b = pd.read_csv('C:/Users/revanth/Desktop/agg_review.csv',low_memory=False)
b.drop(b.columns[b.columns.str.contains('unnamed',case = False)],axis = 1)

merged = a.merge(b, on='business_id')
merged.drop(merged.columns[merged.columns.str.contains('unnamed',case = False)],axis = 1)

merged.to_csv('C:/Users/revanth/Desktop/output.csv', index=False)
```
