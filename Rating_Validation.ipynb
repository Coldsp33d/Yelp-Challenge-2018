```python
reviews_json_df = reviews_json[['business_id','review_id','date','stars']]
reviews_json_df = reviews_json_df.sort_values("business_id")
```

```python
highestrating = reviews_json_df.loc[reviews_json_df['business_id']== "9SK6FWPhLa1q1rjhQd31Rg"]
highestrating = highestrating.sort_values("date")

x = highestrating["date"].values
y = highestrating["stars"].values
plt.plot(x, y,'g--')
plt.xlabel('Dates')
plt.ylabel('Rating for 9SK6FWPhLa1q1rjhQd31Rg')
```

```python
lowestrating = reviews_json_df.loc[reviews_json_df['business_id']== "kE7uP-sybELPfvsEZRzwqg"]
lowestrating = lowestrating.sort_values("date")

a = lowestrating["date"].values
b = lowestrating["stars"].values
plt.plot(a, b,'r--')
plt.xlabel('Dates')
plt.ylabel('Rating for kE7uP-sybELPfvsEZRzwqg')
```
